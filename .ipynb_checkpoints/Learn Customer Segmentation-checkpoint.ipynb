## Implementing Clustering

Clustering algorithms like K-means are sensitive to the scale of the data. Therefore, it's important to standardize the features to ensure they contribute equally to the distance calculations.

```python
from sklearn.preprocessing import StandardScaler
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features = pd.DataFrame(scaled_features, columns = features.columns)
scaled_features
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize = (10,5))
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Before applying K-Means it's important to determine how many cluster we need, there are two ways to find the optimal number of clusters. First, The elbow method. To determine the optimal number of cluster we have to find the "Elbow Point" which is where the rate of decrease slows down.

```python
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_features)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores)
plt.title('Silhouette Scores')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

Second, The silhouette score. This method measures how similar an object is to its own cluster compared to other clusters. A higher silhouette score indicates better-defined clusters.

```python
optimal_clusters = 6
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)
```

```python
data['Cluster'] = cluster_labels
data.head()
```

```python
cluster_means = data.groupby('Cluster').mean()
cluster_means
```

```python
sns.pairplot(
    data,
    hue = 'Cluster',
    vars = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)'],
    palette=sns.color_palette("tab10")
)
plt.suptitle('Pair Plot of Clusters', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(data=data, x='Age', y='Annual Income (k$)', hue='Cluster', palette='tab10')
plt.title('Age vs Annual Income')

plt.subplot(1, 3, 2)
sns.scatterplot(data=data, x='Age', y='Spending Score (1-100)', hue='Cluster', palette='tab10')
plt.title('Age vs Spending Score')

plt.subplot(1, 3, 3)
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='tab10')
plt.title('Annual Income vs Spending Score')

plt.tight_layout()
plt.show()
```
